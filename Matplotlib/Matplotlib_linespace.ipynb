# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 그래프 생성
# x 가 1씩 증가할 때 y는 x*2 씩 증가
flag, ax = plt.subplots()
x = np.arange(5)
y = x * 2
ax.plot(
    x,y,
    linestyle=":",
    marker="*",
    color="blue"
)
```

### 그래프 라인 스타일 변경
* linestyles 종류
* "-" , "--" , "-." , ":"

```python
flag, ax = plt.subplots()
ax.plot(x,x, linestyle="-")
ax.plot(x,x+2, linestyle="--")
ax.plot(x,x+4, linestyle="-.")
ax.plot(x,x+6, linestyle=":")
```

### 라인 컬러 변경

```python
flag, ax = plt.subplots()
ax.plot(x,x, color="g") # green
ax.plot(x,x+2, color="lightblue")
ax.plot(x,x+4, color="0.3") # 검정색 선에 투명도 적용
ax.plot(x,x+6, color="#362FA1")
```

### Marker 설정
* marker 종류
* "." , "o" , "v" , "s" , "*"

```python
flag, ax = plt.subplots()
ax.plot(x,x, marker=".")
ax.plot(x,x+2, marker="o")
ax.plot(x,x+4, marker="v")
ax.plot(x,x+6, marker="s")
ax.plot(x,x+8, marker="*")
```

### 축 경계 조정

```python
# 0 ~ 20 사이의 2000개의 값 생성
a = np.linspace(0,20,2000)
a
```

```python
flag, ax = plt.subplots()
ax.plot(a, np.sin(a))

# x축, y축 보여주는 크기 지정
ax.set_xlim(-2,22)
ax.set_ylim(-1.2,1.2)
```
